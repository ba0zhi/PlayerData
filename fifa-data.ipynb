# Using K-means to label player

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score,calinski_harabasz_score
from sklearn.mixture import GaussianMixture
```

- first, we need to decied the value of K

```python
SSE = []
for each in range(2,14):
    kmean = KMeans(n_clusters = each)
    kmean.fit(temp22)
    SSE.append(kmean.inertia_) # estimator.inertia_获取聚类准则的总和

X = range(2,14)
plt.xlabel('k')
plt.ylabel('SSE')
plt.plot(X,SSE,'o-')
plt.show()
```

```python
Sscore = []
for each in range(2,14):
    kmean = KMeans(n_clusters = each)
    kmean.fit(temp22)
    i_label = kmean.predict(temp22)
    Sscore.append(silhouette_score(temp22, i_label)) # estimator.inertia_获取聚类准则的总和

X = range(2,14)
plt.xlabel('k')
plt.ylabel('S-Score')
plt.plot(X,Sscore,'o-')
plt.show()
```

```python
Cscore = []
for each in range(2,14):
    kmean = KMeans(n_clusters = each)
    kmean.fit(temp22)
    i_label = kmean.predict(temp22)
    Cscore.append(calinski_harabasz_score(temp22, i_label)) # estimator.inertia_获取聚类准则的总和

X = range(2,14)
plt.xlabel('k')
plt.ylabel('S-Score')
plt.plot(X,Cscore,'o-')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
from tensorflow import keras as kr

```

- A list contains player who occured in all 6 records

```python
all_players = []
for each in data_22['ID'].values:
    if ((each in data_21['ID'].values) and 
        (each in data_20['ID'].values) and 
        (each in data_19['ID'].values) and 
        (each in data_18['ID'].values) and 
        (each in data_17['ID'].values)):
        if data_22[data_22['ID'] == each]['Best Position'].values != 'GK':
            all_players.append(each)
len(all_players)
```

```python
cols = []
for each in data_22.drop(columns =['Weak Foot','Skill Moves','Marking','Name', 'Age', 'Photo', 'Nationality', 'Flag','Preferred Foot','Overall',
       'Potential', 'Club', 'Club Logo', 'Value', 'Wage', 'Special', 'International Reputation','Work Rate', 'Body Type', 'Real Face', 'Position',
       'Jersey Number', 'Joined', 'Loaned From', 'Contract Valid Until',
       'Height', 'Weight','Best Position', 'Best Overall Rating','Release Clause','DefensiveAwareness']).columns:
    cols.append(each)

cols
```

- trim the data into six panda frames representing all player data in 6 years

```python
learningData22 = pd.DataFrame([])
learningData22 = PlayerStatus22.drop(columns = ['Weak Foot','Skill Moves','Marking','DefensiveAwareness','GKDiving','GKHandling','GKKicking','GKPositioning','GKReflexes'])

learningData21 = pd.DataFrame([])
learningData21 = PlayerStatus21.drop(columns = ['Weak Foot','Skill Moves','Marking','DefensiveAwareness','GKDiving','GKHandling','GKKicking','GKPositioning','GKReflexes'])

learningData20 = pd.DataFrame([])
learningData20 = PlayerStatus20.drop(columns = ['Weak Foot','Skill Moves','Marking','DefensiveAwareness','GKDiving','GKHandling','GKKicking','GKPositioning','GKReflexes'])

learningData19 = pd.DataFrame([])
learningData19 = PlayerStatus19.drop(columns = ['Weak Foot','Skill Moves','Marking','GKDiving','GKHandling','GKKicking','GKPositioning','GKReflexes'])

learningData18 = pd.DataFrame([])
learningData18 = PlayerStatus18.drop(columns = ['Weak Foot','Skill Moves','Marking','GKDiving','GKHandling','GKKicking','GKPositioning','GKReflexes'])

learningData17 = pd.DataFrame([])
learningData17 = PlayerStatus17.drop(columns = ['Weak Foot','Skill Moves','Marking','GKDiving','GKHandling','GKKicking','GKPositioning','GKReflexes'])

learningDatas = [learningData17, learningData18, learningData19, learningData20, learningData21, learningData22]
```

 - Aggregating data into a 6 by 5189 by 33 numpy array
 - which means 5189 players in 6 years with 33 features

```python
training_players = np.full((len(all_players),6,28), np.nan)
for i in range(len(learningDatas)):
    for j in range(len(all_players)):
        if all_players[j] in learningDatas[i]['ID'].values:
            to_add = learningDatas[i][learningDatas[i]['ID'] == all_players[j]].drop(columns = 'ID').to_numpy()
            training_players[j][i] = to_add
```

```python
np.shape(training_players)
trainX = training_players[0:519*6, 0:-1]
testX = training_players[519*6:,0:-1]
trainLabel = training_players[0:519*6, -1]
testLabel = training_players[519*6:, -1]
np.shape(trainLabel)
```

```python
model = kr.models.Sequential()
model.add(kr.layers.Bidirectional(kr.layers.LSTM(132, 
                                                 return_sequences=True,
                                                 recurrent_dropout=0.2),
                                                input_shape=(trainX.shape[1], trainX.shape[2])))
# model.add(kr.layers.Dropout(0.2))
model.add(kr.layers.Bidirectional(kr.layers.LSTM(66)))
model.add(kr.layers.Dense(28))
model.compile(loss='mae', 
              optimizer='rmsprop')
model.fit(trainX,trainLabel, epochs = 50, batch_size = 32, 
      validation_data =(testX, testLabel))
```
